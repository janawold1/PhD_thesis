# Validating Founder SV's

This data is the result of visual validation of structural variants (SVs) in 2 founder indviduals with samplot. Here, we explore the size distribution and type of SVs prior to trio binning.

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
plt.style.use('seaborn')

RH = pd.read_csv('/mnt/c/Users/Jana/Desktop/kakapo-data/Richard_H_notrio_supported.csv')
WB = pd.read_csv('/mnt/c/Users/Jana/Desktop/kakapo-data/samplot_img/unfiltered_founders/Waynebo/supported/Waynebo_notrio_supported.csv')

#Obtaining the sizes of validated SVs...
RH['size'] = RH['end'] - RH['start']
WB['size'] = WB['end'] - WB['start']

founders = pd.concat(['RH', 'WB'], keys = ['RH', 'WB'])

RH.sort_values(by = ['chr', 'type', 'size'], inplace = True)
WB.sort_values(by = ['chr', 'type', 'size'], inplace = True)

RH_del = RH[RH['type']=="DEL"]
RH_dup = RH[RH['type']=="DUP"]
RH_inv = RH[RH['type']=="INV"]

WB_del = WB[WB['type']=="DEL"]
WB_dup = WB[WB['type']=="DUP"]
WB_inv = WB[WB['type']=="INV"]

display(founders.head())
display(RH.head())
display(WB.head())
```

Now the data is prepped. Here we are plotting the type and size distribution of SVs. As you can see, the majority of SVs are under 1kb in length. 

```python
#Histogram of type of variant
plt.subplot(1, 2, 1)
plt.hist(RH['type'])
plt.title('Structural Variant Type - RH')
plt.ylim((0,350))

plt.subplot(1, 2, 2)
plt.hist(WB['type'])
plt.title('Structural Variant Counts - WB')
plt.ylim((0,350))

plt.tight_layout()
```

```python
plt.subplot(3, 2, 1)
plt.hist(RH_del['size'], bins = 100)
plt.title('Deletions - RH')
plt.ylim((0,125))

plt.subplot(3, 2, 3)
plt.hist(RH_dup['size'], bins = 100)
plt.title('Duplications - RH')
plt.ylim((0,25))

plt.subplot(3, 2, 5)
plt.hist(RH_inv['size'], bins = 100)
plt.title('Inversions - RH')
plt.ylim((0,15))

plt.subplot(3, 2, 2)
plt.hist(WB_del['size'], bins = 100)
plt.title('Deletions - WB')
plt.ylim((0,125))

plt.subplot(3, 2, 4)
plt.hist(WB_dup['size'], bins = 100)
plt.title('Duplications - WB')
plt.ylim((0,25))

plt.subplot(3, 2, 6)
plt.hist(WB_inv['size'], bins = 100)
plt.title('Inversions - WB')
plt.ylim((0,15))

plt.tight_layout()
```

```python
plt.subplot(3, 2, 1)
plt.hist(RH_del['chr'])
plt.title('Deletions per chromosome - RH')
plt.ylim((0,150))

plt.subplot(3, 2, 2)
plt.hist(WB_del['chr'])
plt.title('Deletions per chromosome - WB')
plt.ylim((0,150))

plt.subplot(3, 2, 3)
plt.hist(RH_dup['chr'])
plt.title('Duplications per chromosome - RH')
plt.ylim((0,8))

plt.subplot(3, 2, 4)
plt.hist(WB_dup['chr'])
plt.title('Duplications per chromosome - WB')
plt.ylim((0,8))

plt.subplot(3, 2, 5)
plt.hist(RH_inv['chr'])
plt.title('Inversions per chromosome - RH')
plt.ylim((0,6))

plt.subplot(3, 2, 6)
plt.hist(WB_inv['chr'])
plt.title('Inversions per chromosome - WB')
plt.ylim((0,6))

plt.tight_layout()
```
